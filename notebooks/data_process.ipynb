```python
df_clean= df.dropna(subset = ["cleaned_review"])
```

```python
df_size = df_clean.loc[:, ['cleaned_review', 'sentiments']]
```

```python
df_size.shape
```

```python
df_size["sentiments"].value_counts()
```

```python
df_size["sentiments"] = df_size["sentiments"].replace("negative", 2)
df_size["sentiments"] = df_size["sentiments"].replace("positive", 1)
df_size["sentiments"] = df_size["sentiments"].replace("neutral", 0)
```
